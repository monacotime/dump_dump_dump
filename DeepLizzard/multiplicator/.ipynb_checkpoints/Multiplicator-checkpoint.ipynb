```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
import numpy as np
```

```python
model = Sequential()
```

```python
model.add(Dense(units=64, activation='relu', input_dim=2))
model.add(Dense(units=64, activation='softmax'))
model.add(Dense(units=1, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
data = pd.read_csv("multiples.csv")

x_train = np.array(data[["num1","num2"]])
y_train = np.array(data["results"])
```

```python
print(np.array([0.1,0.1]))
```

```python
model.fit(x_train, y_train, epochs=5, batch_size=32)
```

```python
predition = model.predict(np.array([0.1,0.1]))
```
